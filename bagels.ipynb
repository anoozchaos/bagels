```python
import random

NUM_DIGITS = 3
MAX_GUESSES = 10

def main():
    
    print("Bagels a number guessing game.\nI am thinking of a 3-digit number with no repeated digits.\nTry to guess what it is. Here are some clues:\n\nWhen I say:  That means:\nPico         One digit is correct but in the wrong place.\nFermi        One digit is correct and is in the right position.\nBagels       No digits are correct.\n\nFor example, if the number was 248 and your guess was 486, the clues would be Fermi Pico.\n")
    
    while True: # Main loop
        # This stores the secret number the player needs to guess.
        secretNum = getSecretNum()
        print('I have thought up a number.') 
        print('You have {} guesses to find that number.'.format(MAX_GUESSES))    
          
        numGuesses = 1
        while numGuesses <= MAX_GUESSES:
            guess = ''
            # Keep entering until they enter a valid guess
            while len(guess) != NUM_DIGITS or not guess.isdecimal():
                print('Guess #{}: '.format(numGuesses))
                guess = input('>')

            clues = getClues(guess, secretNum)
            print(clues)
            numGuesses += 1
          
            if guess == secretNum: 
                print('You got it!')
                break  # They're correct, so break out of this loop.
            
        if numGuesses > MAX_GUESSES:
            print('You ran out of guesses.')
            print('The secret number was {}.'.format(secretNum))
          
        # Ask player if they want to play again 
        print('Do you want to play again? (yes or no)')
        if not input('> ').lower().startswith('y'):
            break
          
    print('Thanks for playing!')

def getSecretNum():
    numbers = list('0123456789')  # Creates a list of digits 0 to 9
    random.shuffle(numbers)  # Shuffle them into random order.
          
    # Get the first NUM_DIGITS digit in the list for the secret number.
    secretNum = ''
    for i in range(NUM_DIGITS):
        secretNum += str(numbers[i]) 
    return secretNum
          
def getClues(guess, secretNum):
    "Returns a string with pico, fermi and bagels clues for a guess and a secret number pair."
    if guess == secretNum:
        return 'You got it!'
          
    clues = []
          
    for i in range(len(guess)):
        if guess[i] == secretNum[i]:
            # A correct digit is in a correct place.
            clues.append('Fermi')
        elif guess[i] in secretNum: 
            # A correct digit is in the correct place.
            clues.append('Pico')

    if len(clues) == 0:
        return 'Bagels'  # There are no correct digits at all.
    else:
        # Sort the clues into alphabetical order so their original order doesn't give information away.
        clues.sort()
        # Make a single string from a list of string clues.
        return ' '.join(clues)

# If the program is run (instead of imported), run the game:
if __name__ == '__main__':
    main()

          
          
          
          
```
